## 1 AAMs Basics

### 1.1 Introduction

Active Appearance Models (AAMs) are generative parametric models that describe the shape and appearance of a certain object class; e.g. the human face. In a typical application, these models are matched against input images to obtain the set of parameters that best describe a particular instance of the object being modelled.

The aim of this short notebook is to showcase how one can build and fit AAMs to images using Menpo.

### 1.2 Acquiring data

AAMs are typically build from a large collection of annotated images. In this notebook we will build AAMs from annotated images of the human face and, consequently, this is the object this AAMs will model.

### 1.3 Loading data

The first step in building an AAM is to import all the images and annotations of the training set of LFPW database. Luckly, Menpo's I/O module allows us to do exactly that using only a few lines of code:

```python
import menpo.io as mio

training_images = []
# load landmarked images
for i in mio.import_images(path_to_lfpw + 'lfpw/trainset/*.png'):
    # crop image
    i.crop_to_landmarks_proportion(0.1)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    training_images.append(i)
```

The previous cell loads all the images (with their corresponding landmarks) of the LFPW.

Note that all images are cropped (in order to save some valuable memory space) and converted to grayscale (line 11). The convertionConvertion to rescale is optional (although common typically common in AAMs) and it is done for simplicity.

In order to check if the data has been correctly imported we will define a simple visualization function using the widget package of IPython Notebook. All this function does is to use the `interact` funtion to nicely display the a list of of annotated images.

```python
import matplotlib.pyplot as plt
from IPython.html.widgets import interact

def browse_images(images, group=None, label='all'):
    n = len(images)
    def view_image(i):
        images[i].landmarks[group][label].view()
        plt.show()
    interact(view_image, i=(0,n-1))
```

```python
%matplotlib inline

browse_images(training_images)
```

We can visually check that the data has been correctly imported by simply playing with the slider bar on the previous cell.

### 2.3 Building a simple AAM

Deformable model, in general, and AAMs, in particular, are a core concept in Menpo and all efforts have being put to facilitate their usage. In fact, given a list of training images, an AAM can be build using a single line of code.

```python
from menpo.fitmultilevel.aam import AAMBuilder

# build AAM
aam = AAMBuilder(feature_type=None, normalization_diagonal=100, n_levels=1).build(training_images, verbose=True)
```

As first class citizens of Menpo, AAM can be printed just a other Menpo's objects such as Images or PointClouds:

```python
print aam
```

Printing an AAM is the easiest way to retrieve its particular characteristics. For example, printing the previous AAM tells us that it was build from 811 images among several other things.

AAMs also define an instance method that allows us to generate novel AAM instance by giving particular values to their shape and apperance compoenents:

```python
%matplotlib inline

# using default parameters
aam.instance().view()

# varying shape parameters
aam.instance(shape_weights=[1.2, 0.5, -2.3]).view_new()

# varying appearance parameters
aam.instance(appearance_weights=[2.7, 3.5, 0.9]).view_new()

# varying both
aam.instance(shape_weights=[1.2, 0.5, -2.3], appearance_weights=[2.7, 3.5, 0.9]).view_new()
```

Furthermore, it is very simple to define powerfull visualization functions that allow us to explore AAMs using the previously introduced widget package of the IPython Notebook and the previous AAMs instance method:

```python
def visualize_aam(aam, level=-1, bounds=(-3.0, 3.0), channels=None):    
    def view_aam(shape_pc1, shape_pc2, shape_pc3, tex_pc1, tex_pc2, tex_pc3):
        instance = aam.instance(shape_weights=[shape_pc1, shape_pc2, shape_pc3], 
                                appearance_weights=[tex_pc1, tex_pc2, tex_pc3],
                                level=level)
        instance.landmarks.view(channels=channels)
        plt.show()
    interact(view_aam, shape_pc1=bounds, shape_pc2=bounds, shape_pc3=bounds, tex_pc1=bounds, tex_pc2=bounds, tex_pc3=bounds)
```

```python
%matplotlib inline

visualize_aam(aam)
```

### 2.4 Fit a simple AAM 

In Menpo, AAMs can be fitted to images by creating a `Fitter` object around them. 

One of the most popular and well known family of algorithms for fitting AAMs is the one based on the original `Lucas-Kanade` algorithm for Image Alignment. In order to fit an AAM using algorithms of the previous family Menpo allows you to define a `LucasKanadeAAmFitter` object again using a single line of code.

```python
from menpo.fitmultilevel.aam import LucasKanadeAAMFitter

# define Lucas-Kanade based AAM fitter
fitter = LucasKanadeAAMFitter(aam, n_shape=0.9, n_appearance=0.9)
```

The previous cell has created a `LucasKanadeAAMFitter` that will fit images using 90% of the variance of the AAM shape and appearance models and that will use the default `LucasKanade` algorithm (i.e. the `Alternating Inverse Compositional` algorithm) for fitting. Note that Menpo implements several `LucasKanade` based algorithm for fitting AAMs and that they can be selected by specifying the right keyword arguments on the `LucasKanadeAAMFitter` call.

It is also important to know that these `LucasKanadeAAMFitter` objects are first class citizen in Menpo and, consequently, can be printed:

```python
print fitter
```

Fitting a `LucasKanadeAAMFitter` to an image is as simple as calling its `fit` method. Let's try it by fitting some images of the test set of LFPW database!

```python
import menpo.io as mio

# load test images
test_images = []
for i in mio.import_images(path_to_lfpw + 'lfpw/testset/*.png', max_images=5):
    # crop image
    i.crop_to_landmarks_proportion(0.5)
    # convert it to grayscale if needed
    if i.n_channels == 3:
        i = i.as_greyscale(mode='luminosity')
    # append it to the list
    test_images.append(i)
```

Note that for the purpose of this simple fitting demonstration we will just try to fit the first 5 images of the LFPW test set.

```python
fitting_results = []
# fit images
for j, i in enumerate(test_images):
    # obtain original landmarks
    gt_s = i.landmarks['PTS'].lms
    # generate perturbed landmarks
    s = fitter.perturb_shape(gt_s)
    # fit image
    fr = fitter.fit(i, s, gt_shape=gt_s) 
    # append fitting result to list
    fitting_results.append(fr)
    # print fitting information
    print 'Image: ', j
    print fr
```

Menpo's `Fitter` objects save the result of their fittings using Menpo's `FittingResult` objects. The `LucasKanadeAAMFitter` is no exception and, consequently, the result obtained by executing the previous cell is a list of `FittingResult` objects.

`FittingResult` objects are core Menpo objects that allow us to print, visualize and analyse the results produced by `Fitter` objects. Like all Menpo's core objects `FittingResult` are printable. Note that, they were printed inside of the previous fitting  loop in order to display the final fitting error of each image.

Apart from being printable, `FittingResult` objects also allow us to visualize the initialization from which the fitting procedure started and the obtained final fitting results:

```python
fr = fitting_results[0]
```

```python
%matplotlib inline

# visualize initialization
fr.view_initialization(new_figure=True)

# visualize final fitting result
fr.view_final_fitting(new_figure=True)
```

Furthermore, they allow us to visualize the evolution of the landmark error throughout the fitting process:

```python
%matplotlib inline

fr.plot_error(color_list=['b'], marker_list=['*'])
```

Note that we can visually check the final fitting results for all fitted images by using the previously defined function `browse_images`:

```python
%matplotlib inline

fitted_images = [fr.final_fitting for fr in fitting_results]
browse_images(fitted_images, group='fitted')
```

One of the most powerful and general characteristics of the `AAMBuilder` is that it allows the building of AAMs with arbitrary (user defined) appearance features. Instead of restricting the user to a set of predifine appearance features, the `AAMBuilder` allows him/her to define new features (perhaps combinations of existing image features suported by Menpo or perhaps completely new (even discriminative!) features designed by the user). 

All that is required it that the user encapsulates its feature computation in an function that receives as an arguments a `Menpo` `Image` and that outputs a `Menpo` `Image`:

```python
def my_new_features(image):
    return image.features.igo().features.igo()
```

Although we shouwed how we can build and AAM. I fact menpo is a flexible platform for building aams allowing the building of very different kinds of AAM. We'll stop here and we'll take a look at some of the available options.

```python
aam2 = AAMBuilder(feature_type=my_new_features, normalization_diagonal=100).build(training_images, verbose=True)
```
