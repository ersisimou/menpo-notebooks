# Image Features

This notebook provides a guide on how to extract features from an image. The features that are currently implemented are:

1. [Histogram of Oriented Gradients](http://ieeexplore.ieee.org/xpls/abs_all.jsp?arnumber=1467360&tag=1) (**HOG**)
2. [Image Gradient Orientations](http://ieeexplore.ieee.org/xpl/articleDetails.jsp?arnumber=6138864) (**IGO**)
3. [Edge Structures](http://ieeexplore.ieee.org/xpls/abs_all.jsp?arnumber=990655) (**ES**)
4. [Local Binary Patterns](http://ieeexplore.ieee.org/xpls/abs_all.jsp?arnumber=1017623) (**LBP**)

We also show, in Section 5, how features can be combined.

Let's begin by importing the necessary packages:

# 1. HOG Features

The HOG (Histogram of Oriented Gradients) method clusters gradient orientations in different bins for localized sub-windows of the input image. The resultant feature vector represents a binning of the orientations in to a given number of bins. 

Within Menpo, we have two different methodologies for computing HOGs, **dense** and **sparse**.

### 1.1 Dense HOGs

Dense HOGs are a formulation of HOGs that calculate a feature vector for every pixel in the image. Unlike the original formulation of HOGs (which we refer to as sparse), every single pixel is processed. This is particularly useful for parametric image alignment as it allows you to treat the resulting feature space as an image.

Below is an example of dense HOGs with sampling step of 3 pixels horizontally and vertically. In this example, we include all the possible HOG parameters.

```python
%%time
hog = breaking_bad.features.hog(mode='dense',
                                algorithm='dalaltriggs',
                                cell_size=8,
                                block_size=2,
                                num_bins=9,
                                signed_gradient=True,
                                l2_norm_clip=0.2,
                                window_height=1, window_width=1, window_unit='blocks',
                                window_step_vertical=3, window_step_horizontal=3, window_step_unit='pixels',
                                padding=True,
                                verbose=True,
                                constrain_landmarks=True)
```

The resultant feature image can be viewed in a number of ways, to allow for simpler analysis of the effectiveness of HOGs at capturing the information in the image. The simplest way to visualise HOGs is to render each channel seperately.

```python
%matplotlib inline
hog.view(masked=False);
```

We also provide a 'glyph' view, which plots the gradient vectors (in a circle) and scales their intensity by the vector magnitude. The `vectors_block_size` argument allows you to provide more pixels per circle so that the individual gradients are more easily identifiable.

```python
%matplotlib inline
hog.glyph().landmarks.view();
```

However, if you use a `vectors_block_size` of 1 pixel, you will be able to see the dominant gradients for each pixel.

```python
%matplotlib inline
hog.glyph(vectors_block_size=1,).landmarks.view();
```

### 1.2 Sparse HOGs

Setting the _mode_ option to _sparse_ returns sparsely-sampled HOGs as defined in the original HOGs paper.

```python
%%time
%matplotlib inline
hog = breaking_bad.features.hog(mode='sparse',
                                algorithm='zhuramanan',
                                verbose=True)
hog.glyph(vectors_block_size=4).view();
```

Again, we can plot all of the channels to see what information was captured

```python
%matplotlib inline
hog.view(masked=False);
```

### 1.3 Default Options

Let's take a look at how HOGs perform if we use our default **dense** options.

```python
%%time
hog = breaking_bad.features.hog()
```

These options return the most densely-sampled HOG version and therefore are understandbly slow!

```python
print(hog)
```

They return a HOG image with the same width and height as the input image and 36 channels. Here we see how to render the first 9 of these channels:

```python
%matplotlib inline
hog.view(channels=range(9));
```

The glyph view shows the dominant gradient directions

```python
%matplotlib inline
hog.glyph(vectors_block_size=1).landmarks.view();
```

### 1.4 Constrain Landmarks

In some cases, depending on the options given by the user, the landmarks may end up outside of the bounds of the resultant HOG image. By enabling the flag `constrain_landmarks`, the landmarks that lie outside the borders can be constrained to the image bounds. The default value is `constrain_landmarks=True`. This applies for all feature categories. For example:

```python
%matplotlib inline
# clipping disabled
plt.subplot(121); plt.title('Clipping disabled')
breaking_bad.resize([150, 150]).features.hog(mode='sparse',constrain_landmarks=False).landmarks.view(channels=1,masked=False)
# clipping enabled
plt.subplot(122); plt.title('Clipping enabled')
breaking_bad.resize([150, 150]).features.hog(mode='sparse').landmarks.view(channels=1,masked=False)
```

### 1.5 Windows Centres

All feature types return a matrix with the coordinates of the windows centers on which they were computed:

```python
hog = breaking_bad.features.hog(mode='sparse')
print(hog.pixels.shape)
print(hog.window_centres.shape)
```

# 2. IGO Features

The IGO (Image Gradient Orientations) features concatenate the `cos` and `sin` of the gradient orientation angles for each image pixel. The `cos` and `sin` can be computed either on the orientation angles or on both the angle and double angle of the pixel orientations.

### 2.1 Single angles

Computing the orientation angle is as simple as:

```python
%%time
igo_single = breaking_bad.features.igo()
```

```python
print(igo_single)
```

There are 6 channels total, a `sin` and a `cos` for for `RGB` channel

```python
%matplotlib inline
igo_single.view(channels='all');
```

You can also visualise the glyph view to give you the dominant orientations

```python
%matplotlib inline
igo_single.glyph().view();
```

### 2.2 Double angles

IGOs only have one parameter, the `double_angles` flag. This parameter determines whether or not the double angle is also computed. The double angle is particularly useful for IGOs as it helps mitigate against the ambiguity of the sign of the angle.

```python
%%time
igo = breaking_bad.features.igo(double_angles=True, verbose=True)
```

With `double_angle=True`, there are now 12 channels (the angle and double angle of each of the `RGB` channels).

```python
%matplotlib inline
igo.view(channels='all');
```

The glyph view in the double angle case shows more detail, due to the increase in the number of feature channels.

```python
%matplotlib inline
igo.glyph(vectors_block_size=1).landmarks.view();
```

# 3. ES Features

The ES (Edge Structure) feature provides a measure which captures the orientation of image edge structure at each pixel, together with an indication of how accurate the orientation estimate is. The accuracy belief measure penalizes the orientations in flat, noisy regions and favours the ones near strong edges. Here is an example:

```python
%%time
es = breaking_bad.features.es(verbose=True)
```

```python
%matplotlib inline
es.view(channels=[0, 1]);
```

The glyph view gives an indication of the dominant gradient direction.

```python
%matplotlib inline
es.glyph(vectors_block_size=1).landmarks.view();
```

# 4. LBP Features

The basic idea behind the LBP (Local Binary Patterns) feature is to encode the local structure in an image by comparing each pixels intensity value with its neighbors and then assign an appropriate code. The user can control the density through the `window_step_vertical` and `window_step_horizontal` parameters. At each pixel, the LBP code is computed on circle(s) with radius (radii) defined in `radius`. For each `radius` value, the user has to define the respective number of sampling points, using the parameter `samples`. The parameter `mapping_type` defines the LBP codes mapping: `uniform-2`, `rotation-invariant` or `uniform-2 and rotation_invariant`.

In this example, we extract dense LBP features using two `radius/samples` combinations. The example also illustrates all the possible parameters to the LBP function.

```python
%%time
lbp = breaking_bad.features.lbp(radius=[3, 4],
                                samples=[10, 12],
                                mapping_type='ri',
                                window_step_vertical=1,
                                window_step_horizontal=1,
                                window_step_unit='pixels',
                                padding=True,
                                verbose=True,
                                constrain_landmarks=True)
```

```python
%matplotlib inline
lbp.view(channels='all')
```

The default values for LBPs yield the following results:

```python
%%time
lbp = breaking_bad.features.lbp()
```

```python
%matplotlib inline
lbp.glyph(vectors_block_size=1).landmarks.view();
```

Note that all implemented features hold the parameter values chosen by the user:

```python
print lbp.lbp_parameters
```

### 4.1 LBP Codes Histogram

LBPs are ideal to be used as image descriptors. This is achieved by binning the codes of the output image, as follows:

```python
hist, bin_edges = breaking_bad.as_greyscale().features.lbp(radius=1, samples=8).as_histogram()
```

```python
%matplotlib inline
width = 0.7 * (bin_edges[1] - bin_edges[0])
center = (bin_edges[:-1] + bin_edges[1:]) / 2
plt.bar(center, hist, align='center', width=width);
```

# 5. Features Combination

Features can be used recursively. That is to say, a user is able to extract features from a features image (though this not make much sense!). For example we can compute the LBP features of the IGO-based image:

```python
%%time
lbp_igo = breaking_bad.as_greyscale().features.igo(verbose=True).features.lbp(radius=2, samples=8, mapping_type='none', verbose=True)
```

```python
%matplotlib inline
lbp_igo.landmarks.view(channels='all');
```
